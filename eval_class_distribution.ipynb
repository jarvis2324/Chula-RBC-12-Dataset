```python
import os
```

```python
train = os.listdir('/content/blood_dataset/CBC_BCCD_EythrocytesIDB_2/train')
```

```python
textfiles = []
```

```python
for file in train:
  if(file[-4:]=='.txt'):
    textfiles.append(file)
```

```python
textfiles
```

```python
Platelets = 0
Circular_RBC = 0
WBC = 0
Elgonated_RBC = 0
Other_RBC = 0
```

```python
path = '/content/blood_dataset/CBC_BCCD_EythrocytesIDB_2/train/'
```

```python
for file in textfiles:
  r = open(path + file,'r')
  m = r.readlines()
  for line in m:
    l = line.split()
    if(l[0]=='0'):
      Platelets = Platelets+1
    if(l[0]=='1'):
      Circular_RBC = Circular_RBC+1
    if(l[0]=='2'):
      WBC = WBC+1
    if(l[0]=='3'):
      Elgonated_RBC = Elgonated_RBC+1
    if(l[0]=='4'):
       Other_RBC = Other_RBC+1
```
